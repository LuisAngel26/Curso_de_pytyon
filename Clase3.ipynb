```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
indice22=pd.read_csv("https://raw.githubusercontent.com/jamc88/Analisis-datos-atmosfericos/refs/heads/main/datos/indice_2022.csv",parse_dates=True, index_col=0)
indice23=pd.read_csv("https://raw.githubusercontent.com/jamc88/Analisis-datos-atmosfericos/refs/heads/main/datos/indice_2023.csv",parse_dates=True, index_col=0)

```

```python
indice23.head(25)
```

```python
len(indice22)
```

```python
indice=pd.concat([indice22,indice23],axis=0)
```

```python
indice
```

```python
ozono=indice[['Noroeste ozono', 'Noreste ozono','Centro ozono','Suroeste ozono','Sureste ozono']]
```

```python
ozono
```

```python
ozono.info()
```

```python
ozono.isna().sum()
```

```python
ozono=ozono.fillna({'Noroeste ozono': ozono['Noroeste ozono'].median(),
             'Noreste ozono': ozono['Noreste ozono'].median(),
                          'Centro ozono': ozono['Centro ozono'].median(),
                                       'Suroeste ozono': ozono['Suroeste ozono'].median(),
                                                    'Sureste ozono': ozono['Sureste ozono'].median()})
```

```python
ozono
```

```python
ozono=ozono.resample('D').mean()
```

```python
ozono
```

```python
fig,ax=plt.subplots(figsize=(10,5),dpi=200)
ax.plot(ozono['Noroeste ozono'].resample('M').mean(),label='Noroeste ozono',lw=1,color='violet',marker='o',markersize=6);
ax.plot(ozono['Noreste ozono'].resample('M').mean(),label='Noreste ozono',lw=1,color='red',marker='o',markersize=6);
ax.plot(ozono['Centro ozono'].resample('M').mean(),label='Centro ozono',lw=1,color='yellow',marker='o',markersize=6);
ax.plot(ozono['Suroeste ozono'].resample('M').mean(),label='Suroeste ozono',lw=1,color='blue',marker='o',markersize=6);
ax.plot(ozono['Sureste ozono'].resample('M').mean(),label='Sureste ozono',lw=1,color='green',marker='o',markersize=6);

ax.tick_params(labelsize=20)
ax.tick_params(labelsize=10)
ax.set_xlabel('Tiempo',fontdict={'fontsize':12})
ax.set_ylabel('O3 ($\mu$g/$m^3)',fontdict={'fontsize':12})
ax.legend(prop={'size':9})
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/jamc88/Analisis-datos-atmosfericos/refs/heads/main/datos/rama_2023_05.csv", parse_dates=True, index_col=0)
```

```python
import pymannkendall as mk
```

```python
mk_O3=mk.original_test(df['O3'],alpha=0.05)
```

```python
mk_O3
```

```python
a=mk_O3.slope
b=mk_O3.intercept
```

```python
def f5(x):
  return x*a+b
```

```python
x = range(0, 3073)
fig, ax = plt.subplots(figsize=(10,5),dpi=200)
ax.plot(x, df['O3'],lw=1,color='black');
ax.plot(x, [f5(i) for i in x], color='red',lw=4)
```
